```python
import csv  # loads csv module
with open("fruits1.csv", 'r', newline = '', encoding = 'utf-8') as file:
    s = csv.reader(file)         # reader() function takes each
                                 # row (lines) into a list
    for row in s:                #iterable olduğu için liste oldu
        print(row)
```

```python
#  dosya sonu karakterler  newline = ''  "\r\n"
# csv okurken newline "" olursa satır sonu
# delimiter ayraç default , virgüle ayarlı
```

```python
import csv

with open("fruits.csv", 'r', newline = '', encoding = 'utf-8') as file:
    csv_rows = csv.reader(file, delimiter= ",")  
                        # gives the same output as the previous one                          
    for row in csv_rows:     # delimiter ,
        print(row) 
```

```python
import csv

with open("fruits.csv", 'r', encoding = 'utf-8') as file:
    csv_rows = csv.reader(file,delimiter= ":")    # : dan ayıracaktı bulamadı
                                                # gives the same output as the previous one          
    for row in csv_rows: 
        print(row) 
```
